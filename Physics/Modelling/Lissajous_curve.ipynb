# **–ú–æ–¥–µ–ª–∏—Ä–æ–≤–∞–Ω–∏–µ –Ω–∞ —Ç–µ–º—É "–§–∏–≥—É—Ä—ã –õ–∏—Å—Å–∞–∂—É"**


> **–§–∏–≥—É—Ä—ã –õ–∏—Å—Å–∞–∂—É** –ø–æ–ª—É—á–∞—é—Ç—Å—è –ø—Ä–∏ —Å–ª–æ–∂–µ–Ω–∏–∏ –¥–≤—É—Ö –≤–∑–∞–∏–º–Ω–æ –ø–µ—Ä–ø–µ–Ω–¥–∏–∫—É–ª—è—Ä–Ω—ã—Ö –∫–æ–ª–µ–±–∞–Ω–∏–π —Ä–∞–∑–ª–∏—á–Ω–æ–π —á–∞—Å—Ç–æ—Ç—ã. –û—Å–Ω–æ–≤–Ω–∞—è –æ–±–ª–∞—Å—Ç—å –ø—Ä–∏–º–µ–Ω–µ–Ω–∏—è —Ñ–∏–≥—É—Ä –õ–∏—Å—Å–∞–∂—É - —Ä–∞–¥–∏–æ—Ç–µ—Ö–Ω–∏–∫–∞.


–§–∏–≥—É—Ä—ã –õ–∏—Å—Å–∞–∂—É –ø–æ–ª—É—á–∞—é—Ç—Å—è –ø—Ä–∏ —Å–ª–æ–∂–µ–Ω–∏–∏ –¥–≤—É—Ö –≤–∑–∞–∏–º–Ω–æ –ø–µ—Ä–ø–µ–Ω–¥–∏–∫—É–ª—è—Ä–Ω—ã—Ö –∫–æ–ª–µ–±–∞–Ω–∏–π —Ä–∞–∑–ª–∏—á–Ω–æ–π —á–∞—Å—Ç–æ—Ç—ã:
$x=ùëã_mcosœâ_1t; \quad y=ùëå_mcos(œâ_2t+\phi)$
–†–∞—Å—Å–º–æ—Ç—Ä–∏–º —Å–ª—É—á–∞–π, –∫–æ–≥–¥–∞ —á–∞—Å—Ç–æ—Ç—ã –æ–¥–∏–Ω–∞–∫–æ–≤—ã $œâ_1$=$œâ_2$. –ò—Å–∫–ª—é—á–∞—è –≤—Ä–µ–º—è –∏–∑ —É—Ä–∞–≤–Ω–µ–Ω–∏–π, –ø–æ–ª—É—á–∞–µ–º —É—Ä–∞–≤–Ω–µ–Ω–∏–µ —Ç—Ä–∞–µ–∫—Ç–æ—Ä–∏–∏ –≤ —è–≤–Ω–æ–º –≤–∏–¥–µ:
$(\frac{x}{ùëã_m})^2+(\frac{y}{ùëå_m})^2-\frac{2xy}{ùëã_mùëå_m}cosœï=sin^2œï$



> –¢–∞–∫–∂–µ, –¥–ª—è –Ω–∞—à–µ–≥–æ —É–¥–æ–±—Å—Ç–≤–∞ –µ—Å—Ç—å –ø—Ä–µ–¥—Å—Ç–∞–≤–ª–µ–Ω–∏–µ –ú–∞—Ç–µ–º–∞—Ç–∏–µ—á—Å–∫–æ–≥–æ –≤—ã—Ä–∞–∂–µ–Ω–∏–µ—è –¥–ª—è —Ñ–∏–≥—É—Ä (–∫—Ä–∏–≤—ã—Ö) –õ–∏—Å—Å–∞–∂—É:
$\begin{cases}
x(t)=Asin(at+Œ¥)
\\y(t)=Bsin(bt)
\end{cases}$
, –≥–¥–µ A,B - –∞–º–ø–ª–∏—Ç—É–¥—ã –∫–æ–ª–µ–±–∞–Ω–∏–π, $a,b$ - —á–∞—Å—Ç–æ—Ç—ã, $Œ¥$ - —Å–¥–≤–∏–≥ —Ñ–∞–∑.


–í–∏–¥ –∫—Ä–∏–≤–æ–π —Å–∏–ª—å–Ω–æ –∑–∞–≤–∏—Å–∏—Ç –æ—Ç —Å–æ–æ—Ç–Ω–æ—à–µ–Ω–∏—è $a/b$. –ö–æ–≥–¥–∞ —Å–æ–æ—Ç–Ω–æ—à–µ–Ω–∏–µ —Ä–∞–≤–Ω–æ 1, —Ñ–∏–≥—É—Ä–∞ –õ–∏—Å—Å–∞–∂—É –∏–º–µ–µ—Ç –≤–∏–¥ —ç–ª–ª–∏–ø—Å–∞, –ø—Ä–∏ –æ–ø—Ä–µ–¥–µ–ª–µ–Ω–Ω—ã—Ö —É—Å–ª–æ–≤–∏—è—Ö –æ–Ω–∞ –∏–º–µ–µ—Ç –≤–∏–¥ –æ–∫—Ä—É–∂–Ω–æ—Å—Ç–∏ ($A=B,Œ¥=œÄ/2$ —Ä–∞–¥–∏–∞–Ω) –∏ –æ—Ç—Ä–µ–∑–∫–∞ –ø—Ä—è–º–æ–π ($Œ¥ = 0$). –ï—â–µ –æ–¥–∏–Ω –ø—Ä–∏–º–µ—Ä —Ñ–∏–≥—É—Ä—ã –õ–∏—Å—Å–∞–∂—É - –ø–∞—Ä–∞–±–æ–ª–∞ ($a/b=2, Œ¥=œÄ/2$). –ü—Ä–∏ –¥—Ä—É–≥–∏—Ö —Å–æ–æ—Ç–Ω–æ—à–µ–Ω–∏—è—Ö —Ñ–∏–≥—É—Ä—ã –õ–∏—Å—Å–∞–∂—É –ø—Ä–µ–¥—Å—Ç–∞–≤–ª—è—é—Ç —Å–æ–±–æ–π –±–æ–ª–µ–µ —Å–ª–æ–∂–Ω—ã–µ —Ñ–∏–≥—É—Ä—ã, –∫–æ—Ç–æ—Ä—ã–µ —è–≤–ª—è—é—Ç—Å—è –∑–∞–º–∫–Ω—É—Ç—ã–º–∏ –ø—Ä–∏ —É—Å–ª–æ–≤–∏–∏ $a/b$ - —Ä–∞—Ü–∏–æ–Ω–∞–ª—å–Ω–æ–µ —á–∏—Å–ª–æ. 

**–°–Ω–∏–∑—É –ø—Ä–µ–¥—Å—Ç–∞–≤–ª–µ–Ω—ã —Ñ–∏–≥—É—Ä—ã –õ–∏—Å–∞–∂—É –ø—Ä–∏ –æ–ø—Ä–µ–¥–µ–ª–µ–Ω–Ω—ã—Ö –ø—Ä–æ–ø–æ—Ä—Ü–∏—è—Ö –∏ –æ–ø—Ä–µ–¥–µ–ª–Ω–Ω–æ–π "Œ¥"**


Screenshot.png

**–ü–æ–ª—É—á–µ–Ω–∏–µ –∏–∑ —Å–∏—Å—Ç–µ–º—ã –Ω–∞—à–∏—Ö —É—Ä–∞–≤–Ω–µ–Ω–∏–π —É—Ä–∞–≤–Ω–µ–Ω–∏–µ —Ç—Ä–∞–µ–∫—Ç–æ—Ä–∏–∏ –¥–ª—è –ø—Ä–æ–∏–∑–≤–æ–ª—å–Ω—ã—Ö —á–∞—Å—Ç–æ—Ç**
$\begin{cases}
x=Acos(œât)
\\y=Bcos(œât+Œ±)
\end{cases}$
, –≥–¥–µ Œ± - —Ä–∞–∑–Ω–æ—Å—Ç—å —Ñ–∞–∑ –æ–±–æ–∏—Ö –∫–æ–ª–µ–±–∞–Ω–∏–π,  –ê –∏ –í —Ä–∞–≤–Ω—ã –∞–º–ø–ª–∏—Ç—É–¥–∞–º —Å–∫–ª–∞–¥—ã–≤–∞–µ–º—ã—Ö –∫–æ–ª–µ–±–∞–Ω–∏–π. –£—Ä–∞–≤–Ω–µ–Ω–∏–µ —Ç—Ä–∞–∫–µ—Ç–æ—Ä–∏–µ —Ä–µ–∑—É–ª—å—Ç–∏—Ä—É—é—â–µ–≥–æ –∫–æ–ª–µ–±–∞–Ω–∏—è –æ–ø—Ä–µ–¥–µ–ª–∏–º –∏—Å–∫–ª—é—á–µ–Ω–∏–µ–º –∏–∑ —Å–∏—Å—Ç–µ–º—ã –≤—Ä–µ–º–µ–Ω–∏ t. –ó–∞–ø–∏—à–µ–º —Å–∫–ª–∞–¥—ã–≤–∞–µ–º—ã–µ –∫–æ–ª–µ–±–∞–Ω–∏—è:

$\frac{x}{A}=cos(wt)
\\\frac{y}{B}=cos(wt+Œ±)= cos(wt)cos(Œ±)-sin(wt)sin(Œ±)
$

–ó–∞–º–µ–Ω–∏–º –≤–æ –≤—Ç–æ—Ä–æ–º —É—Ä–∞–≤–Ω–µ–Ω–∏–∏ $cos(wt)$ –Ω–∞ $\frac{x}{A}$ –∏ $sin(wt)$ –Ω–∞ $\sqrt{1-\frac{x}{A}^2}$. 

–ü–æ—Å–ª–µ —ç—Ç–æ–≥–æ –Ω–∞–π–¥–µ–º –ø–æ–ª—É—á–∏–≤—à–∏–µ—Å—è —É—Ä–∞–≤–Ω–µ–Ω–∏–µ(–ø–æ–ª—É—á–∏–ª–æ—Å—å —É—Ä–∞–≤–Ω–µ–Ω–µ–∏–µ —ç–ª–ª–∏–ø—Å–∞), —É –∫–æ—Ç–æ—Ä–æ–≥–æ –æ—Å–∏ –æ—Ä–∏–µ—Ç–∏—Ä–æ–≤–∞–Ω—ã –ø—Ä–æ–∏–∑–≤–æ–ª—å–Ω–æ –æ—Ç–Ω–æ—Å–∏—Ç–µ–ª—å–Ω–æ –∫–æ–æ—Ä–¥–∏–Ω–∞—Ç–Ω—ã—Ö –æ—Å–µ–π:

$\frac{x^2}{A^2}-\frac{2xy}{AB}cos(Œ±)+\frac{y^2}{B^2}=sin^2(Œ±)$

**–ù–∏–∂–µ –ø—Ä–∏–≤–µ–¥–µ–Ω –∞–ª–≥–æ—Ä–∏—Ç–º –ø–æ–ª—É—á–µ–Ω–∏—è —Ñ–∏–≥—É—Ä –õ–∏—Å—Å–∞–∂—É, –ø—Ä–∏ –∑–∞–ø—É—Å–∫–µ –∫–æ—Ç–æ—Ä–æ–≥–æ –º–æ–∂–Ω–æ –ø–æ–ª—É—á–∏—Ç—å –ª—é–±—ã–µ –Ω—É–∂–Ω—ã–µ –Ω–∞–º —Ñ–∏–≥—É—Ä—ã, –∏–∑–º–µ–Ω—è—è –ø–∞—Ä–∞–º–µ—Ç—Ä—ã $Œ±,Œ≤$ –∏ $Œ¥$**

```python
import math
import matplotlib
import numpy
from ipywidgets import interactive, RadioButtons


def lissajous_curve(Œ±, Œ≤, Œ¥):
  matplotlib.pyplot.figure()
  t = numpy.linspace(0, 10, 1000)
  x = numpy.sin(Œ± * t + Œ¥)
  y = numpy.sin(Œ≤ * t)
  matplotlib.pyplot.plot(x, y)
  matplotlib.pyplot.axis("equal")
  matplotlib.pyplot.axis("off")
  matplotlib.pyplot.show()

interactive_plot = interactive(
    lissajous_curve, 
    Œ± = (1, 10), 
    Œ≤ = (1, 10), 
    Œ¥ = RadioButtons(options=[('0', 0), ('œÄ / 4', (math.pi/4)), 
                              ('œÄ / 2', (math.pi/2)), ('3œÄ / 4', (3*math.pi)/4), 
                              ('œÄ', math.pi)], description = 'Œ¥ ')
  )
interactive_plot
```
